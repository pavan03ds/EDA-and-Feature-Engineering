# Bi-Variate and Multi-Variate Analysis

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
tips = sns.load_dataset('tips')
```

```
tips.sample(10)
```

```
tips.info()
```

```
flights = sns.load_dataset('flights')
```

```
flights.sample(10)
```

```
flights.info()
```

```
iris = pd.read_csv('Iris.csv')
```

```
iris.sample(10)
```

```
iris.info()
```

```
titanic = pd.read_csv('titanic.csv')
titanic.sample(10)
```

```
titanic.info()
```

### Scatterplot (Numerical - Numerical)

```
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size')
plt.show()
```

### Bar plot (numerical - categorical)

```
titanic.groupby('Pclass')['Age'].sum().plot(kind='bar')
```

```
plt.bar(titanic['Pclass'], titanic['Age'])  #using matplotlib
```

```
sns.barplot(x='Pclass', y='Age', data=titanic, hue='Sex')  #using seaborn
```

### Box Plot (numerical - categorical)

```
sns.boxplot(x='Sex', y='Age', data=titanic, hue='Survived')
```

### Distplot (numerical - categorical)

```
sns.displot(x='Age', data=titanic[titanic['Survived']==0], color='Red')
sns.displot(x='Age', data=titanic[titanic['Survived']==1], color='Blue')
plt.show()
```

```
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False, color='Red')
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
plt.show()
```

## Heat Map (categorical - categorical)

```
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic['Survived']))
```

```
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind="bar")
```

```
titanic.groupby('Sex')['Survived'].mean()*100
```

```
(titanic.groupby('Embarked')['Survived'].mean()*100)
```

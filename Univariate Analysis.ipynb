```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
df = pd.read_csv('titanic.csv')
```

```
#how does my data look
df.head()
```

```
#sample rows
df.sample(5)
```

```
#data type of columns
df.info()
```

```
#are there any missing values
df.isna().sum()
```

```
#descriptive statistics
df.describe()
```

```
#are there any duplicate values
df.duplicated().sum()
```

```
#how is the correlation between columns
df.corr()['Survived']
```

# Univariate Analysis

## Categorical Data

###Count plot

```
sns.countplot(x=df['Survived'])
#df.Survived.value_counts().plot(kind='bar')
plt.show()
```

###pie chart

```
df.Survived.value_counts().plot(kind='pie',autopct='%.2f')
plt.show()
```

## Numerical Data

###Histogram

```
sns.histplot(x=df['Age'], bins=10)  #using seaborn
plt.show()
```

```
plt.hist(df['Age'], bins=10) #using matplotlib
plt.show()
```

- Distplot

```
sns.distplot(x=df['Age'])  #warning is that `distplot` is a deprecated function and will be removed in seaborn v0.14.0.

plt.show()           #we have to use `displot` or `histplot' in place of distplot
```

###Boxplot

```
sns.boxplot(x=df['Fare'])
plt.show()
```

```
sns.boxplot(x='Age', data=df)
plt.show()
```
